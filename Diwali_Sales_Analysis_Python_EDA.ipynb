** Gender **

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
yx=sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From the above graphs we can see that most of the purchases and Amount spent are by Female

```python
df.columns
```

** Age Group **

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
agex=sns.barplot(x='Age Group',y='Amount',data=age_sales)
```

From the above graphs we can see that Highest number of purchases and Amount spent are from the Age Group 26 - 35

```python
df.columns
```

** State **

```python
state_order=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=state_order)
```

```python
salesamt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=salesamt)
```

From the above graphs we can see that though Kerala has highest count of Orders the Amount spent is less

** Marital Status **

```python
sns.set(rc={'figure.figsize':(5,5)})
mx=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in mx.containers:
    mx.bar_label(bars)
```

```python
marsales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=marsales,hue='Gender')
```

From the above graphs we can see that most of the buyers are married (women).

** OCCUPATION**

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ox=sns.countplot(x='Occupation',data=df)
for bars in ox.containers:
    ox.bar_label(bars)
```

From the above graph we can see that IT sector has the highest number(1583) of Orders, second being Healthcare and third Aviation

```python
occsales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=occsales,hue='Amount')
```

From the above graph we see that TOP 3 buyers are working in IT Sector, Healthcare and AViation

```python
sns.set(rc={'figure.figsize':(15,5)})
occsales=df.groupby(['Occupation','Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(30)
sns.barplot(x='Occupation',y='Amount',data=occsales,hue='Product_Category')
```

From the above graph we see the Product Category on which each Occupation holders have spent money. Like IT Sector have spent most of their money on Food followed by Footwear & shoes, Clothing and Gadgets.

** Product Category **

```python
sns.set(rc={'figure.figsize':(20,5)})
pcx=sns.countplot(x='Product_Category',data=df)
for bars in pcx.containers:
    pcx.bar_label(bars)
```

```python
pc_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=pc_sales,hue='Product_Category')
```

From the above graph we can see that Food is the highest sold Product followed by Clothing & Apparel.

** Product ID **

```python
pi_sales=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=pi_sales)
```

From the above graph we can see that Product_ID P00265242 is the highest sold Product.
